```python
import pandas as pd

df = pd.read_csv('/content/Fish.csv')
```

```python
import numpy as np
```

```python
df = df.iloc[:34]
```

```python
x = df['Height']
y = df['Weight']
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.scatter(x,y,label = 'train', alpha=0.5)
```

```python
w = np.random.randn(1)
b = np.random.randn(1)
```

```python
fig = plt.figure(figsize=(6,6))

plt.scatter(x,y)
plt.plot(x,b+w*x,color = 'red')
plt.show()
```

```python
def mse(w,b,data): ##오차구하기
  yhat = w*x+b   ##(예측한값)
  error = y-yhat  ##(실제값=y)
  loss = (error**2).mean()
  return error, loss
```

```python
def sgd(lr, x, x_grad):
  result = x-lr*x_grad
  return result
```

```python
def grad(error, data):
  w_grad = -2*(x*error).mean()
  b_grad = -2*error.mean()
  return w_grad, b_grad
```

```python
lr = 0.001        ## 한번에 얼마나 차이를 둘껀지
n_epochs = 100000  ##반복시키는 횟수
error_loss = []
for epoch in range(n_epochs):
  error, loss = mse(w,b,x)
  error_loss.append(loss)
  w_grad, b_grad = grad(error,x)
  w = sgd(lr,w,w_grad) ##업데이트한 기울기
  b = sgd(lr,b,b_grad) ##업데이트한 절편
print(w,b)
```

```python
plt.figure(figsize=(25,10))
#plt.yscale('log')
plt.scatter([i for i in range(n_epochs)], error_loss, s=5, c='r',label='loss')
```

```python
fig = plt.figure(figsize=(6,6))

plt.scatter(x,y)
plt.plot(x,b+w*x,color = 'red')
plt.show()
```
